```python
from pandas import read_csv
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from keras import optimizers
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Activation


train_data = pd.read_csv("train.csv")
train_data.head()

test_data = pd.read_csv("test.csv")
# test_data.head()
```

```python
# print(train_data.isnull().sum())
# print('\n')
# print(test_data.isnull().sum())
```

```python
# Finding and fixing Null/NaN Values in Embarked and Fare for Train and Test set

# Train
train_data.groupby('Embarked').nunique() # Q have lowes number of people 
train_data['Embarked'] = train_data['Embarked'].replace(np.nan, 'Q')

# Test 
test_data.describe()
test_data['Fare'] = test_data['Fare'].replace(np.nan, 35)
```

```python
# print(train_data.isnull().sum())
# print('\n')
# print(test_data.isnull().sum())
```

```python
# Filling Train data columns with encoded values
df_train = train_data.drop(["Name","Age","Cabin","Ticket"],axis = 1)
df_train['Sex'] = df_train['Sex'].map( {'female': 0, 'male': 1} ).astype(int)
df_train["Embarked"] = df_train['Embarked'].map( {'S':0, 'C':1, 'Q':2} ).astype(int)
# df_train
```

```python
# Filling Test data columns with encoded values
df_test = test_data.drop(["Name","Age","Cabin","Ticket"],axis = 1)
df_test['Sex'] = df_test['Sex'].map( {'female': 0, 'male': 1} ).astype(int)
df_test["Embarked"] = df_test['Embarked'].map( {'S':0, 'C':1, 'Q':2} ).astype(int)
# df_test
```

```python
df_train_input = df_train.drop(columns=['Survived'])
df_train_result = df_train["Survived"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_train_input, df_train_result, test_size=0.1)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
# Learning rate
learning_rate = 0.01

# Epoch  
epochs = 400

# Optimizer
opt = optimizers.Adam(lr=learning_rate)
```

```python
# NN Model

model = Sequential()
model.add(Dense(50, activation='relu', input_dim = 7))
model.add(Dense(40, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='BinaryCrossentropy', optimizer=opt, metrics=['accuracy'])
```

```python
# Summary
# model.summary()
```

```python
# Model Training

history = model.fit(
    x_train,
    y_train,
    epochs=epochs,
    verbose = 0
)
```

```python
df_test_result = model.predict(x_test)
```

```python
test_result = []
for i in df_test_result:
    if i < 0.5:
        test_result.append(0)
    else:
        test_result.append(1)
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test, test_result))
```
