```python
from pandas import read_csv
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from keras import optimizers
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import Activation
from pandas import get_dummies
from sklearn.preprocessing import MinMaxScaler


train_data = pd.read_csv("train.csv")
train_data.head()

test_data = pd.read_csv("test.csv")
# test_data.head()
```

```python
train_data.head()
test_data.head()
```

```python
train_data.isnull().sum()
# train_data["Age"].count()
```

```python
test_data.isnull().sum()
```

```python
# Filling Train values

train_data['Embarked'] = train_data['Embarked'].replace(np.nan, 'C')
train_data["Age"].fillna(train_data["Age"].mean(),inplace=True)
```

```python
# Filling Test values

test_data['Fare'] = test_data['Fare'].replace(np.nan, 35)
test_data["Age"].fillna(test_data["Age"].mean(),inplace=True)
```

```python
# Dropping unwanted colums

df_train = train_data.drop(["Name","Age","Cabin","Ticket"],axis = 1)
df_test = test_data.drop(["Name","Age","Cabin","Ticket"],axis = 1)
```

```python
df_train.shape
```

```python
df_test.shape
```

```python
# Converting Both Train and test colums to Categorial data to use df.dummies()
df_train["Sex"] = df_train["Sex"].astype('category')
df_train["Embarked"] = df_train["Embarked"].astype('category')
df_test["Sex"] = df_test["Sex"].astype('category')
df_test["Embarked"] = df_test["Embarked"].astype('category')
```

```python
# Using df.dummies to one hot encode the Sex and Embarked columns

df_train = pd.get_dummies(df_train, columns=["Sex","Embarked"])
df_test = pd.get_dummies(df_test, columns=["Sex","Embarked"])
```

```python
# Scaling values in Fare column
mms = MinMaxScaler()
df_train[['Fare']] = mms.fit_transform(df_train[['Fare']])
df_test[['Fare']] = mms.fit_transform(df_test[['Fare']])
```

```python
df_train_input = df_train.drop(columns=['Survived'])
df_train_result = df_train["Survived"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_train_input, df_train_result, test_size=0.1)
```

```python
# Learning rate
learning_rate = 0.01

# Epoch  
epochs = 50

# Optimizer
opt = optimizers.Adam(lr=learning_rate)
```

```python
# NN Model

model = Sequential()
# model.add(Dropout(0.2, input_shape=(9,)))
model.add(Dense(10, activation='relu', input_dim = 10))
model.add(Dense(1, activation='sigmoid'))
```

```python
# Complier

model.compile(loss='BinaryCrossentropy', optimizer=opt, metrics=['accuracy'])
```

```python
# Model Training

history = model.fit(
    x_train,
    y_train,
    epochs=epochs,
    verbose = 0
)
```

```python
df_test_result = model.predict(x_test)
```

```python
test_result = []
for i in df_test_result:
    if i < 0.5:
        test_result.append(0)
    else:
        test_result.append(1)

```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test, test_result))
```
